```python
crop_dict = {
    'rice': 1,
    'maize': 2,
    'jute': 3,
    'cotton': 4,
    'coconut': 5,
    'papaya': 6,
    'orange': 7,
    'apple': 8,
    'muskmelon': 9,
    'watermelon': 10,
    'grapes': 11,
    'mango': 12,
    'banana': 13,
    'pomegranate': 14,
    'lentil': 15,
    'blackgram': 16,
    'mungbean': 17,
    'mothbeans': 18,
    'pigeonpeas': 19,
    'kidneybeans': 20,
    'chickpea': 21,
    'coffee': 22
}
data['crop_num']=data['label'].map(crop_dict)
```

```python
X = data.drop(['crop_num','label'],axis=1)
Y = data['crop_num']
X.shape,Y.shape
```

## correlation analysis

```python
corr=X.corr()
corr
```

```python
import seaborn as sns
sns.heatmap(corr, annot =True , cbar= True)
```

```python
#Since K has high correlation, we remove it 
X=X.drop(['K'],axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state =0)
```

```python
X_train.shape, X_test.shape,Y_train.shape,Y_test.shape
```

```python
X_train.head()
```

## L1 normalization

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer(norm='l1')
X_train= norm.fit_transform(X_train)
X_test=norm.fit_transform(X_test)
```

## Standard scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

## Classification models with their accuracy

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
gmb=GaussianNB()
rfc=RandomForestClassifier()
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
rfc.fit(X_train,Y_train)
ypred = rfc.predict(X_test)
print("Accuracy of Random Forest : ",str(np.round(accuracy_score(Y_test,ypred)*100,2)) + '%')
```

```python
gmb.fit(X_train,Y_train)
ypred = gmb.predict(X_test)
print("Accuracy of Gausian Naive Bayes : ",str(np.round(accuracy_score(Y_test,ypred)*100,2)) + '%')
```
